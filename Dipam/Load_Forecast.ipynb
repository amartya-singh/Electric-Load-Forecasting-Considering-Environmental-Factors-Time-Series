```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
prev = 15
forecast = 5
hours = 24

previous_hrs = prev * hours 
forecasting_hrs = forecast * hours
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/AI_ML_Projects/Electric_Load_Forecasting_Considering_Environmental_Factors_Time_Series_Demo_Project/Dipam/data/continuous_dataset.csv')
dataset['datetime'] = pd.to_datetime(dataset['datetime'])
print(dataset.head())
```

```python
true_forecast = pd.read_csv('/content/drive/MyDrive/AI_ML_Projects/Electric_Load_Forecasting_Considering_Environmental_Factors_Time_Series_Demo_Project/Dipam/data/weekly_pre_dispatch_forecast.csv')
true_forecast['datetime'] = pd.to_datetime(true_forecast['datetime'])
print(true_forecast.head())
```

```python
data = pd.merge(dataset, true_forecast, how="inner", on="datetime")
print(data.head())
```

```python
data.info()
```

```python
load_fore = data['load_forecast']
df = data.drop(columns = ['datetime','load_forecast'])
df.info()
print(load_fore)
```

```python
X_train = df.values
X_test = df.iloc[-(previous_hrs + forecasting_hrs):,:]
print(X_train)
print(X_test)
```

```python
y_origin = X_test[-forecasting_hrs:]
print(y_origin)
x_test = X_test[:-forecasting_hrs]
```

```python
print(x_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
scaler = MinMaxScaler(feature_range = (0, 1))
x_train = scaler.fit_transform(X_train)
x_test = scaler.fit_transform(x_test)
y_origin = scaler.fit_transform(y_origin)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_origin.shape)
```

```python
x_test = x_test.reshape((1,x_test.shape[0], x_test.shape[1]))
y_test = y_origin.reshape((1,y_origin.shape[0], y_origin.shape[1]))
```

```python
print(x_test.shape)
print(y_test.shape)
```

```python
from numpy import array
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense
from keras.layers import RepeatVector
from keras.layers import TimeDistributed
from keras.layers import Dropout

 
# split a univariate sequence into samples
def split_sequence(sequence, n_steps_in, n_steps_out):
	X, y = list(), list()
	for i in range(len(sequence)):
		# find the end of this pattern
		end_ix = i + n_steps_in
		out_end_ix = end_ix + n_steps_out
		# check if we are beyond the sequence
		if out_end_ix > len(sequence):
			break
		# gather input and output parts of the pattern
		seq_x, seq_y = sequence[i:end_ix], sequence[end_ix:out_end_ix]
		X.append(seq_x)
		y.append(seq_y)
	return array(X), array(y)
 
# define input sequence
#raw_seq = [10, 20, 30, 40, 50, 60, 70, 80, 90]
# choose a number of time steps
n_steps_in, n_steps_out = previous_hrs, forecasting_hrs
# split into samples
X, y = split_sequence(x_train, n_steps_in, n_steps_out)
# reshape from [samples, timesteps] into [samples, timesteps, features]
n_features = 16
print(X.shape)
print(y.shape)
X = X.reshape((X.shape[0], X.shape[1], X.shape[2]))# n_features))
y = y.reshape((y.shape[0], y.shape[1], y.shape[2]))#n_features))
print(X.shape)
print(y.shape)
# define model
model = Sequential()
model.add(LSTM(100, activation='relu', input_shape=(n_steps_in, n_features),dropout = 0.2))
model.add(RepeatVector(n_steps_out))
model.add(LSTM(100, activation='relu', return_sequences=True,dropout = 0.2))
model.add(TimeDistributed(Dense(y.shape[2])))
model.compile(optimizer='adam', loss='mean_squared_error')
# fit model
stats = model.fit(X, y, epochs=100,batch_size= 1024, validation_split=0.2, validation_batch_size=1024)
```

```python
model.summary()
```

```python
print(stats.history.keys())

plt.plot(stats.history['loss'])
plt.plot(stats.history['val_loss'])
plt.title('loss graph')
plt.ylabel('loss')
plt.xlabel('number of epoch')
plt.legend(['training_loss', 'validation_loss'])
plt.show()
```

```python
predicted_values_t = []
for index , values in enumerate(yhat):
    predicted_values_t.append(scaler.inverse_transform(yhat[index]))

Y_test_t = []
for index , values in enumerate(y_test):
    Y_test_t.append(scaler.inverse_transform(y_test[index]))

print(predicted_values_t[0].shape)
print(Y_test_t[0].shape)
Y_test = np.array(Y_test_t)
predicted_values = np.array(predicted_values_t)
print(Y_test.shape)
print(predicted_values.shape)
```

```python
print(Y_test.shape)
print(predicted_values.shape)

mean_squared_error_array=[]
for index in range(Y_test.shape[0]):
    mean_squared_error_array.append(mean_squared_error(Y_test[index,:,0:1],predicted_values[index,:,0:1]))

root_mean_squared_error_array = np.sqrt(np.array(mean_squared_error_array))

print("Median Root Mean Squared Error: ",np.median(root_mean_squared_error_array), "MWh i.e. MegaWattHour")
print("Accuracy Based on Median RMSE ",100-((np.median(root_mean_squared_error_array)/np.max(Y_test))*100),"%")

print("Max Root Mean Squared Error: ",np.max(root_mean_squared_error_array), "MWh i.e. MegaWattHour")
print("Accuracy Based on Max RMSE ",100-((np.max(root_mean_squared_error_array)/np.max(Y_test))*100),"%")
```
